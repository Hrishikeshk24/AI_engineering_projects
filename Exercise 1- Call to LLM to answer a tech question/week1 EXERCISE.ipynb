**Call to gpt-5-nano**



```python
def assess_code(path):
    response = openai.chat.completions.create(
        model=MODEL,
        messages=[
            {"role": "system", "content": link_system_prompt},
            {"role": "user", "content": get_script_user_prompt(path)}
        ]#,
        #response_format={"type": "markdown"}
    )
    result = response.choices[0].message.content
    #return result
    display(Markdown(result))
```

**Call to LLM to answer a technical question**

```python
link_system_new_prompt = """
You are a technical expert who can help answer any technical question around software development, tools, libraries, etc.
You are provided with a question.
You are to answer the question based on your expertise.         
You should respond in markdown format. The response should be concise and to the point.  

# Answer to the question


"""
```

```python
question = """  
Explain tokenization in NLP.
"""
```

**Call to llama3.2:1b**

```python
response = ollama.chat.completions.create(
        model=MODEL_LLAMA,
        messages=[
            {"role": "system", "content": link_system_new_prompt},
            {"role": "user", "content": question}
    ]    )
result = response.choices[0].message.content
display(Markdown(result))
```

```python
response = openai.chat.completions.create(
        model=MODEL_GPT,
        messages=[
            {"role": "system", "content": link_system_new_prompt},
            {"role": "user", "content": question}
    ]    )
result = response.choices[0].message.content
display(Markdown(result))
```

**Using stream**

```python
stream = openai.chat.completions.create(
        model=MODEL_GPT,
        messages=[
            {"role": "system", "content": link_system_new_prompt},
            {"role": "user", "content": question}
          ],
        stream=True
    )    
response = ""
display_handle = display(Markdown(""), display_id=True)
for chunk in stream:
    response += chunk.choices[0].delta.content or ''
    update_display(Markdown(response), display_id=display_handle.display_id)
```
